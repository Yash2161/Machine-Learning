## Multiple Linear Regression

Linear regression with multiple variables

y ~ x1, x2, x3......., xn

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 : Read the dataset

```python
import pandas as pd
df = pd.read_csv('50_Startups.csv')
df.head()
```

PROFIT ~ RND, ADMIN, MKT

PROFIT_pred = B0 + B1 x RND + B2 x ADMIN + B3 x MKT

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 2: Separate X(Independent Variable: RND,ADMIN,MKT) and Y(Dependent Variable: PROFIT)

```python
X = df[['RND', 'ADMIN', 'MKT']]
X.head()
```

```python
Y = df[['PROFIT']]
Y.head()
```

### Step 3: Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
```

```python
model.intercept_
```

```python
model.coef_
```

1. PROFIT_pred = 50122.19 + 0.8057xRND -0.0258xADMIN + 0.0272xMKT
2. Therefore Profit is directly dependent on RND 'cause ADMIN and MKT does not have that much impact on Profit

### Step 3: Evaluate The Model

```python
# Get R2 Score
model.score(X, Y)
```

```python
ypred = model.predict(X)
ypred[0:5]
```

```python
Y.head(
    
)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y, ypred)
mse
```

```python
rmse = mse**(1/2)
rmse
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y, ypred)
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(Y, ypred)
mape
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y, ypred)
r2
```

#### R2 > 0.8 So Good Model

## Regression plot for each column

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for i in ['RND', 'ADMIN', 'MKT']:
    plt.figure(figsize=(10,5))
    sns.regplot(data = df, x=i, y='PROFIT', line_kws={'color':'red'})
    plt.title(f'Regression Plot for {i} vs PROFIT')
    plt.show()
    
```
